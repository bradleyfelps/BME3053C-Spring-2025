# What is GitHub Copilot?

GitHub Copilot is an AI pair programmer that helps you write code faster and with less effort. It works by:

- Suggesting whole lines or blocks of code as you type
- Converting comments into working code
- Helping you understand and work with unfamiliar libraries
- Offering multiple suggestions for you to choose from
- Learning from your coding style and preferences

# Switching AI Models in Copilot

GitHub Copilot offers different AI models to help with various coding tasks. Here's how to switch between them:

## Available Models

- **Default (GPT-3.5 Turbo)**: The standard model for general coding assistance
- **GPT-4**: More powerful model with enhanced code understanding and generation capabilities
- **Copilot Chat**: Conversational interface for complex queries and explanations

## How to Switch Models

1. **In VS Code**:
    - Click on the Copilot icon in the status bar
    - Select "GitHub Copilot: Switch Model"
    - Choose your preferred model from the dropdown menu

2. **Using Command Palette**:
    - Press `Ctrl+Shift+P` (Windows/Linux) or `Cmd+Shift+P` (Mac)
    - Type "GitHub Copilot: Switch Model"
    - Select the desired model

3. **In Copilot Chat**:
    - Type `/model` followed by the model name
    - Example: `/model gpt-4`

## When to Use Different Models

- **Default Model**: Everyday coding tasks, quick suggestions
- **GPT-4**: Complex algorithms, thorough code explanations, challenging problems
- **Specialized Models**: Domain-specific coding (when available)

Remember that more powerful models might be slower to respond or have usage limitations based on your subscription plan.

### ✏️ Exercise Create a function that processes medical image data:
   - Accept a numpy array representing an image
   - Apply basic image processing (normalization, filtering)
   - Return the processed image
   - Include proper error handling

````python
import numpy as np
from scipy.ndimage import gaussian_filter

def process_medical_image(image):
    """
    Processes a medical image by normalizing and applying a Gaussian filter.

    Parameters:
        image (numpy.ndarray): Input image as a numpy array.

    Returns:
        numpy.ndarray: Processed image.

    Raises:
        ValueError: If the input is not a numpy array.
    """
    try:
        # Check if input is a numpy array
        if not isinstance(image, np.ndarray):
            raise ValueError("Input must be a numpy array.")

        # Normalize the image (scale pixel values to range [0, 1])
        normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))

        # Apply a Gaussian filter for smoothing
        filtered_image = gaussian_filter(normalized_image, sigma=1)
        return filtered_image

    except Exception as e:
        print(f"An error occurred while processing the image: {e}")
        return None

# Example usage
if __name__ == "__main__":
    # Create a sample image (random 2D numpy array)
    sample_image = np.random.rand(256, 256) * 255

    # Process the image
    processed_image = process_medical_image(sample_image)

    if processed_image is not None:
        print("Image processed successfully!")
````


### ✏️ Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    """
    A class to manage patient records with methods for adding, updating,
    retrieving records, and performing basic statistical analysis.
    """

    def __init__(self):
        # Initialize an empty dictionary to store patient records
        self.records = {}

    def add_record(self, patient_id, data):
        """
        Adds a new patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            data (dict): Patient data (e.g., name, age, medical history).

        Raises:
            ValueError: If patient_id already exists or data is invalid.
        """
        if patient_id in self.records:
            raise ValueError(f"Patient ID {patient_id} already exists.")
        if not isinstance(data, dict) or "name" not in data or "age" not in data:
            raise ValueError("Invalid data. Must include 'name' and 'age'.")
        self.records[patient_id] = data

    def update_record(self, patient_id, updated_data):
        """
        Updates an existing patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            updated_data (dict): Updated patient data.

        Raises:
            ValueError: If patient_id does not exist or updated_data is invalid.
        """
        if patient_id not in self.records:
            raise ValueError(f"Patient ID {patient_id} does not exist.")
        if not isinstance(updated_data, dict):
            raise ValueError("Invalid data format. Must be a dictionary.")
        self.records[patient_id].update(updated_data)

    def get_record(self, patient_id):
        """
        Retrieves a patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.

        Returns:
            dict: Patient data.

        Raises:
            ValueError: If patient_id does not exist.
        """
        if patient_id not in self.records:
            raise ValueError(f"Patient ID {patient_id} does not exist.")
        return self.records[patient_id]

    def calculate_average_age(self):
        """
        Calculates the average age of all patients.

        Returns:
            float: Average age of patients, or None if no records exist.
        """
        if not self.records:
            return None
        ages = [data["age"] for data in self.records.values() if "age" in data]
        return sum(ages) / len(ages) if ages else None

    def __str__(self):
        """
        Returns a string representation of all patient records.
        """
        return f"Patient Records: {self.records}"


# Example usage
if __name__ == "__main__":
    manager = PatientRecords()

    # Add records
    manager.add_record("P001", {"name": "Alice", "age": 30, "history": ["flu", "cold"]})
    manager.add_record("P002", {"name": "Bob", "age": 40, "history": ["diabetes"]})

    # Update a record
    manager.update_record("P001", {"age": 31})

    # Retrieve a record
    print(manager.get_record("P001"))

    # Calculate average age
    print(f"Average Age: {manager.calculate_average_age()}")

    # Print all records
    print(manager)
````

### ✏️ Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

class DataVisualizationDashboard:
    """
    A class to create a data visualization dashboard with interactive elements.
    """

    def __init__(self, data):
        """
        Initializes the dashboard with a pandas DataFrame.

        Parameters:
            data (pd.DataFrame): The dataset to visualize.
        """
        if not isinstance(data, pd.DataFrame):
            raise ValueError("Data must be a pandas DataFrame.")
        self.data = data

    def plot_histogram(self, column):
        """
        Plots a histogram for a specified column.

        Parameters:
            column (str): The column to plot.
        """
        if column not in self.data.columns:
            raise ValueError(f"Column '{column}' not found in the dataset.")
        plt.hist(self.data[column], bins=20, color='skyblue', edgecolor='black')
        plt.title(f"Histogram of {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.show()

    def plot_scatter(self, x_column, y_column):
        """
        Plots a scatter plot for two specified columns.

        Parameters:
            x_column (str): The column for the x-axis.
            y_column (str): The column for the y-axis.
        """
        if x_column not in self.data.columns or y_column not in self.data.columns:
            raise ValueError(f"Columns '{x_column}' or '{y_column}' not found in the dataset.")
        plt.scatter(self.data[x_column], self.data[y_column], color='purple', alpha=0.7)
        plt.title(f"Scatter Plot: {x_column} vs {y_column}")
        plt.xlabel(x_column)
        plt.ylabel(y_column)
        plt.show()

    def interactive_line_plot(self, column):
        """
        Creates an interactive line plot with a slider to adjust the smoothing factor.

        Parameters:
            column (str): The column to plot.
        """
        if column not in self.data.columns:
            raise ValueError(f"Column '{column}' not found in the dataset.")

        # Prepare the data
        y = self.data[column]
        x = range(len(y))

        # Create the figure and axis
        fig, ax = plt.subplots()
        plt.subplots_adjust(bottom=0.25)
        line, = ax.plot(x, y, label=column)
        ax.set_title(f"Interactive Line Plot: {column}")
        ax.set_xlabel("Index")
        ax.set_ylabel(column)

        # Add a slider for smoothing
        ax_slider = plt.axes([0.2, 0.1, 0.65, 0.03])
        slider = Slider(ax_slider, 'Smoothing', 1, 20, valinit=1, valstep=1)

        # Update function for the slider
        def update(val):
            smoothing = int(slider.val)
            smoothed_y = y.rolling(window=smoothing, min_periods=1).mean()
            line.set_ydata(smoothed_y)
            fig.canvas.draw_idle()

        slider.on_changed(update)
        plt.legend()
        plt.show()


# Example usage
if __name__ == "__main__":
    # Create a sample dataset
    data = pd.DataFrame({
        "Age": [25, 30, 35, 40, 45, 50, 55, 60],
        "Height": [160, 165, 170, 175, 180, 185, 190, 195],
        "Weight": [50, 55, 60, 65, 70, 75, 80, 85]
    })

    # Initialize the dashboard
    dashboard = DataVisualizationDashboard(data)

    # Plot a histogram
    dashboard.plot_histogram("Age")

    # Plot a scatter plot
    dashboard.plot_scatter("Height", "Weight")

    # Create an interactive line plot
    dashboard.interactive_line_plot("Weight")
````
